## Question 1: What is the gender distribution of the survey participants?

Understanding the gender distribution helps us determine whether the responses are balanced across different gender groups.

```python
gender_count=health_df.Gender.value_counts()
gender_count
```

```python
plt.figure(figsize=(10, 5))
plt.title('Gender Distribution')
plt.pie(gender_count,
        labels=gender_count.index,
        autopct='%1.1f%%',
        startangle=180);
```


### Observation

The pie chart illustrates the proportion of survey participants across different gender categories. It helps identify which gender has the highest representation in the dataset.The highest representaion in Menatl Health cases is of Females among all genders.

## Question 2: Which countries contributed the highest number of representation?

To answer this question, we calculate the number of participants from each country using the `value_counts()` function. The results are displayed using a bar chart, making it easy to compare the participation levels across different countries.

```python
country_count=health_df.Country.value_counts()
country_count
```

```python
sns.barplot(x=country_count.index,y=country_count.values)
plt.title("Mental Health Cases Across Countries");
```

### Observation

The bar chart shows the number of survey participants from each country. It reveals that the dataset is dominated by responses from a few countries, while several countries have comparatively fewer participants.Highest Mental Health cases is seen in USA

## Question 3: What is the employment status distribution?

The `value_counts()` function is used to calculate the percentage of participants in each employment category. Displaying these percentages in a horizontal bar chart provides a clear comparison of employment status among respondents.

```python
employement_status_pct=health_df.Employment_Status.value_counts()*100/health_df.Employment_Status.count()
employement_status_pct
```

```python
sns.barplot(x=employement_status_pct,y=employement_status_pct.index)
plt.title("Percentage of Employment Status");
```

### Observation

The chart shows the percentage of participants belonging to different employment categories. It provides an overview of the occupational background of the respondents, helping us understand the diversity of the surveyed population.The menat health cases can be seen among Full-Time employees.

## Question 4: Is there a relationship between sleep duration and daily caffeine consumption?

A scatter plot is used to examine the relationship between the number of hours participants sleep each night and the number of caffeinated drinks they consume daily. The points are also colored according to whether participants experience sleep-related problems.

```python
sns.scatterplot(x=health_df.Sleep_Hours_Night,y=health_df.Caffeine_Drinks_Day,hue=health_df.Sleep_Trouble)
plt.title("Caffeine Consumption vs Sleep Hours");
```

### Observation

The scatter plot helps visualize whether caffeine consumption affects sleep duration. The color coding also allows us to compare participants with and without sleep troubles, making it easier to identify any noticeable trends or clusters.
